## Changing matplotlib plot features in a Jupyter notebook with Python 2.7

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)
fig = plt.gcf()
```

## Note what *does not* work

```python
fig.show()
```
