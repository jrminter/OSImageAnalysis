# Changing matplotlib plots in Jupyter Notebook 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)
fig = plt.gcf()
```

### Note what *does not* work

```python
fig.show()
```

### Note what *does* work

```python
fig
```

### Let's make a change

```python
ax = fig.gca()
ax.set_xlim(left=-1, right=1)
fig
```
