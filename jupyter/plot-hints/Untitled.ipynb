```python
%matplotlib inline
import os
from PIL import Image as pil
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread, imsave
from scipy.misc import face

qual = 85
cSize = 140
```

```python
f = face(gray=True)
f = f[230:230+cSize, 220:220+cSize]
plt.imshow(f, cmap='gray', interpolation='nearest');
plt.show();
```

```python
jpgPth = './f.jpg'
pim = pil.fromarray(np.uint8(f))
pim.save(jpgPth, "JPEG", quality=qual)

j = imread(jpgPth)
print(j.shape, j.dtype)
```

```python
plt.imshow(j, cmap='gray', interpolation='nearest');
plt.show();
```

```python
f = f.astype(float)
j = j.astype(float)
dif = np.abs(f - j)

minV = np.min(dif)
maxV = np.max(dif)

print("Gray level difference - Min: ", minV, ", Max:", maxV)

plt.imshow(dif, cmap='viridis', interpolation='nearest');
plt.show();
```
